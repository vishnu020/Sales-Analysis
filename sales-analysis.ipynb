```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_excel('superstore_sales.xlsx')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
broughts=df.groupby('product_name').agg({'quantity':'sum'}).sort_values('quantity',ascending=False)[:10]
broughts
```

```python
sns.barplot(x='product_name',y='quantity',data=broughts.reset_index())
plt.xticks(rotation = 'vertical')
```

```python
productSales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
productSales.sort_values(by=['sales'], inplace=True, ascending=False)
productSales.head(10)
```

```python
profit_product=pd.DataFrame(df.groupby('product_name').sum()['profit'])
profit_product.sort_values(by=['profit'],inplace=True,ascending=False)
profit_product.head(10)
```

```python
sns.barplot(x='product_name',y='profit',data=profit_product.head(10).reset_index())
plt.xticks(rotation='vertical')
```

```python
df.groupby(['category','sales']).sum()
```

```python
profitable_category=pd.DataFrame(df.groupby('category').sum()['profit'])
profitable_category.sort_values(by=['profit'],inplace=True,ascending=False)
profitable_category
```

```python
sns.barplot(x='category',y='profit',data=profitable_category.reset_index())
```

```python
plt.figure(figsize=(17,5))
plt.bar(df['category'],df['sales'],color = 'b')
plt.xticks(size=8)
plt.show()
```

```python
df.groupby(['category','sub_category']).sum()
```

```python
df.groupby(['category','sub_category'],as_index=False)['quantity'].count()
```

```python
segment_profit=pd.DataFrame(df.groupby('segment').sum()['profit'])
segment_profit.sort_values(by=['profit'],inplace=True,ascending=False)
segment_profit
```

```python
sns.barplot(x='segment',y='profit',data=segment_profit.reset_index())
plt.xticks(rotation='vertical')
```

```python
df.groupby('ship_mode').sum()['quantity']
```

```python
plt.figure(figsize=(17,8))
plt.bar(df['ship_mode'],df['sales'],color='b')
plt.xticks(rotation='vertical')
plt.show
```

```python
category_hist=sns.FacetGrid(df,col='ship_mode',palette='rainbow')
category_hist.map(plt.hist,'category')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='market',data=df,palette='rainbow')
```

```python
country_sales=pd.DataFrame(df.groupby('country').sum()['sales'])
country_sales.sort_values(by=['sales'],inplace=True,ascending=False)
country_sales.head(10)
```

```python
df.groupby('country').agg({'shipping_cost':'mean'}).sort_values('shipping_cost',ascending=False).head(10)
```

```python
data10=df.sort_values(by='profit',ascending=False).head(10)[['order_id','customer_name','profit']]
data10
```

```python
month = df['order_date'].astype(str) 
month = month.apply(lambda x:datetime.strptime(x,'%Y-%m-%d'))
df['month'] = month.map(lambda x: x.month) 
df
```

```python
date = df.groupby(['year','month']).sum()
date
```

```python
sns.barplot(x='year',y='sales',data=date.reset_index())
plt.xticks(rotation = 'vertical')
```
